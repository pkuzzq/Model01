提高收入和生活水平（即人均 GDP）的唯一途径是提高实际生产率，即每单位投入（即每工作小时）产生的产出量

Productivity The Office of Productivity and Technology (OPT) measures how efficiently the U.S. converts inputs into the outputs of goods and services. Measures of labor productivity compare the growth in output to the growth in hours worked and measures of total factor productivity (TFP), also known as multifactor productivity (MFP), compare growth in output to the growth in a combination of inputs that include labor, capital, energy, materials, and purchased services.

您提到的生产力和技术办公室（OPT）是一个关注如何将各种投入转换为产出的效率的机构。这里面涉及到两种重要的生产率衡量方式：劳动生产率和全要素生产率（TFP），也称为多要素生产率（MFP）。让我们详细了解一下这两种衡量方式：

劳动生产率是衡量单位劳动输入（如工作小时数）所产出的商品和服务数量的指标。这是一个相对简单的衡量标准，通常用于评估工人的效率和经济增长的一个方面。计算劳动生产率的公式通常是： $$ \text{劳动生产率} = \frac{\text{总产出}}{\text{劳动投入}} $$ 例如，如果一个工厂的员工在一周内生产了1000个单位的产品，并且总共工作了500小时，那么劳动生产率为2个单位/小时。

全要素生产率（TFP），或称多要素生产率（MFP），是一种更为复杂的衡量标准，它考虑了所有类型的投入（包括劳动、资本、能源、材料和购买的服务）在产出中的综合效率。TFP的计算不仅反映了劳动力的效率，还反映了资本的效率以及技术进步的贡献。TFP的增长通常被视为技术变革和提高效率的指标。它的计算涉及到比较产出的增长与所有这些投入的增长： $$ \text{TFP增长} = \text{产出增长} - \text{加权的投入增长} $$ 举个例子，如果一家公司的产出增长了10%，而劳动力、资本和其他投入的加权平均增长为8%，那么TFP增长为2%。

这两种生产率的衡量对于政策制定者、企业管理层以及经济学家来说都非常重要。劳动生产率的提高通常与工人培训、工作效率改进等因素相关，而TFP的增长则通常被解释为技术创新和更有效的资源管理。国家和企业通过监控这些指标可以评估他们的经济表现、竞争力以及潜在的经济增长。

通过理解和分析这些生产率指标，可以更好地把握经济的健康状况和发展方向，为制定提高经济效率和增长的战略提供数据支持。

## 3 一个主要的前古典观点：重商主义

### 3.1 托马斯·穆勒

他认为，一个国家的财富可以通过积极的贸易平衡来增加，即出口的价值应超过进口的价值。这种思想主张通过对外贸易获取黄金和银子，以此增加国家的财富和宝藏。

经济模型影响：孟的理论强调了国家干预经济活动的重要性，尤其是通过关税和出口补贴来支持本国产业的发展。这种思想主张国家之间的经济竞争，认为贸易是零和游戏。

### 3.2 亚当·斯密

分工显著提升了劳动生产力，是技能、灵巧性和判断力在各地发挥作用的核心。每个人都有一种改善自身状况的强烈动力，这种动力足以推动社会向富裕和繁荣发展，克服各种法律障碍。除了和平、低税负和公正的司法管理外，几乎没有其他方式能将一个国家从落后状态提升至富裕的高度：一切富裕都是自然进程的结果。消费是生产的唯一目标；生产者应专注于在必要的范围内促进消费者的利益。节俭能够增加资本，而挥霍及不当行为会削减资本。一个人节省下的任何收入都会增加其资本，既可以用来支持更多的生产性劳动，也可以通过借贷他人以获得利息来实现这一点。个人的资本只能通过节省年收入来增加……因此，社会上所有个体的资本增长也依赖于同样的方式。一个国家的土地和劳动产出增长只能通过增加生产性劳动者的数量或提高现有员工的工作效率来实现。显然，生产性劳动者的数量不可能大幅增加，但可以通过增加投资或维持他们所需的经费来实现目标。只有通过优化流程、缩短时间和使用合适的工具，或者合理配置工作，才能提高同样数量员工的效率。在这两种情况下，都需要额外的资金投入。

### 3.3 李嘉图

1.一个国家的富裕程度并不取决于它的货币多少或商品的货币价值多高，而是取决于其商品的实际丰富程度——这些商品能够为人们的生活带来舒适和享受。

2.在李嘉图看来，土地是一个固定要素（对于 19 世纪的英国来说肯定有道理），报酬递减首先在大多数（农业）生产性土地上开始耕种将耕种扩展到更差的边际土地和更差的边际土地。当一个地区的所有肥沃土地都已经开垦之后，如果人口和资本继续增长，就需要更多的食物，这些食物只能依赖那些条件较差(越来越边缘的土地)的土地来生产。这就意味着，尽管劳动工资未变，运输产品到消费地的成本增加了，因为需要更多的劳动力和运输工具，这也意味着需要更多的资本来维持同等水平的产品输出。

3.在李嘉图的视角中，如果一个国家的土地面积无限且质量均匀，那么使用这些土地将不会产生任何成本。地租的存在是因为，随着人们开始开垦，需要投入更多的劳动和资本来获取较少的回报。最终，土地的边际产量会减少到零，这意味着没有更多的收益可以从这些土地上获得。

4.长期看来，土地边际产出越来越少，导致资本的利润降低，工人工资下降到仅足以维持基本生活。在这种均衡状态下，地主因为土地稀缺而能够收取高额地租，从而获得巨大收益。他们的收益与社会其他成员的利益形成对立。李嘉图的分析不仅关注在这种均衡状态下地主、资本家和工人的收入分配。在土地报酬递减、经济地租（“李嘉图地租”）——高于机会成本的收益，由于相对质量差异和某些固定因素的稀缺而产生之下，会形成“稳态经济”模式，但衍生出一个政治问题就是“寻租”的现象。

亚当·斯密对土地探讨： 当土地变成私有财产后，地主们就像其他所有者一样，期望在他们未曾耕作的地方收获，还要求为这些自然资源的产出支付租金。

当这些资源属于公共财产时，劳动者只需要付出采集的劳动即可。例如，森林中的木材、田野中的草，以及地球上的所有自然果实，当这些资源属于公共财产时，劳动者只需要付出采集的劳动即可。

当这些资源是私有时，劳动者不仅需要付出采集的劳动，还必须支付一定的费用获取采集许可，且需要将他的劳动所得的一部分上交给地主。这部分所得的价值就构成了土地的租金。因此，土地租金作为一种使用费，实质上是一种垄断价格。这个价格并不基于地主对土地的改良投入或其成本，而是基于农民的承受能力来确定的。这种定价方式体现了土地私有制下的经济压力和社会不平等。

从李嘉图的土地极差地租到经济租。在完全竞争市场中，企业将在 $AC_{min}$ 上组织生产。但如果每个企业的 $MC$ 是不同的（事实上也不可能相同），完全竞争会是生产有效的么？企业为什么非要在 $p^{*}=AC_{min}$ 时进入市场？一些企业效率更高，所以他们进入市场的边际成本和平均成本更低。更低的原因可能是企业家的卓越管理能力或者拥有更优质的资源条件。但是，这些企业一定是非常少见的，就像球员中 LeBron James 等，很难大量出现。这些优质企业，位于行业的 $AC_{min}$ 上，那是因为最低点对应的市场价格 $p$，足够高，以至于可以让效率低的企业获得足够多的收入，维持他们的生存（这个生存方式，可能仅仅够支付可变成本的）。经济租的定义就是，高效率企业获得的收益与其支付的成本（为留住高品质生产要素的支付的机会成本，通常更高）之间的差额，这部分的差额是成本低于行业其他企业平均成本的额外回报，具有正义性。握有经济租的企业是优质企业，他们可以把资源或要素转售给其他企业，比如球员转会、迁址等，那么这个时候的经济租就是一种“成本”。如果把这个经济租的成本与其他成本加总在一起，那么在均衡时，总成本应该等于总收益；除此之外的其他任何位置上，总成本应该大于总收益。若果包含了经济租在内的总成本曲线，推导出的平均成本与企业相切与均衡点，这个 $AC^{\prime}$ 在普通 $AC$ 之上，这样在计算每个企业的规模报酬时候，仍然可以考虑不同价格切于 $AC$ 最低点的情况。

但是这样的处理方式 也有一个缺点，就是资本、劳动等传统要素的会计核算方式与经济租是不同的。经济租是价格决定的，而不是决定价格的因素。在企业生产决策时，考虑的边际成本曲线中并不包含经济租。经济租的概念是生产要素（如土地、劳动力、资本）获得的超过其机会成本的收入。在生产决策中，企业通常将经济租视为固定成本，因此它不随生产量的变化而变化。因此，在边际成本曲线中，经济租通常不包含在内。另一方面，经济租之所以是由价格决定的，是因为如果行业的供给曲线向上倾斜，需求提高就会抬高市场均衡价格，进而使得现有企业在保持产量不变的情况下，收入增加，进而经济租增加。这会引发单个企业的平均成本曲线从$AC^{\prime}$ 再次提高到$AC^{\prime\prime}$，并与新的市场均衡价格$p^{\prime}$ 再次相切。再一次强调，在$MC=MR$的决策中，刻画成本曲线是帮助确定价格接受厂商的生产决策，进而找到市场的均衡价格和均衡数量，由于竞争机制导致经济租趋于 0，因此每个企业都在排除了经济租之外的平均成本最低处进行生产。

### 5.2 简单Solow模型

我们看到各国在储蓄和投资率方面存在巨大差异，（例如）近年来美国投资占 GDP 的 20%，中国占 30%，印度占 12%（宾夕法尼亚大学世界表中实际投资与实际 GDP 的比率）。这些差异对各国的收入和增长率有多重要？答案：不太重要。为什么？因为资本的回报递减率意味着（实际上）额外的资本会产生越来越少的产出。这有点令人失望，但仍然很关键，因为它告诉我们，如果我们想解释我们在各国之间看到的巨大差异，就需要另辟蹊径。这一见解来自 1987 年诺贝尔经济学奖获得者罗伯特·索洛。

在索洛-斯旺经济中，代理人会储蓄他们当前收入的一定比例。储蓄维持或增加了资本存量。资本与劳动力相结合产生产出，而产出又支付给工人和资本所有者。为了简单起见，我们**忽略人口和生产率的增长**。

1.  **生产函数：产出源自（全要素）生产率、资本和/或劳动力的变化**：生产函数是 Cobb-Douglas 形式，对于每个单一生产要素而言，边际产量是下降的。实物资本K能够进行积累。技术增长率是外生的，被一个固定的参数决定。储蓄率和投资率是固定的。该形式能够更好的刻画边际产出递减的特性的生产函数，公式为： $$
    Y_t = A_t K_t^\alpha L_t^{1-\alpha}
    $$ 其中，$Y_t$ 是产出，$K_t$ 是资本存量，$L_t$ 是劳动力，$A_t$ 是全要素生产率（技术水平），$\alpha$ 是资本的产出弹性。该式反映了产出的变化源自（全要素）生产率、资本和/或劳动力的变化。回想一下，该生产函数满足了对资本的递减回报：每一单位的额外资本都会导致产出的较小增加。

2.  **投资全部来源于储蓄**：流量恒等式 $S_t = I_t + (G_t − T_t) + NX_t$ 将家庭或个人储蓄（收入部分）与投资、政府赤字和净出口（支出部分）联系起来。索洛排除了最后两个（可以将它们放回去），因此得到： $$S_t = I_t$$ 这背后隐藏了一个收入与支出恒等式：$Y_t = C_t + I_t$。如何刻画储蓄行为？简单地把储蓄看做是产出（收入）的一部分，并且假设储蓄率是**外生**的且为$s$: $$S_t = sY_t$$ 其中，$s$ 是储蓄率，介于 0 和 1 之间。

3.  **资本存量会有折旧**：资本存量以固定折旧率 $\delta$ ，因此$t+1$期资本存量包含两个部分，一部分是$t$期的投资，另一部分是$t$期资本存量排除掉折旧（损坏）的部分： $$K_{t+1} =  I_t + (1-\delta)K_t
              =  sA_t K_t^\alpha L_t^{1-\alpha}+(1-\delta)K_t$$

4.  **资本存量的动态**。 $$\Delta K_{t+1}=K_{t+1}-K_t=s A K_t^\alpha L^{1-\alpha}-\delta K_t$$ 注意上式中假设**没有人口增长和全要素生产率的增长**，劳动力 $L_t$ 是一个常数 $L$，全要素生产率 $A_t$ 是一个常数 $A$。定义一个新的变量 $k_t$，表示 t 期的每单位劳动力的资本存量，即 $k_t:=K_t / L$。由于生产函数具有齐次性质，将 $K_t$ 和 $L_t$ 同时缩放到 $k_t$ 和 1，生产函数也同时缩放到 $f(k):=F(k, 1)$。 最后，得到了资本动态方程的最终表达式： $$k_{t+1} = g(k) =sf(k)+(1-\delta) k$$ 这个方程描述了资本存量 $k_t$ 在时间上的演变，是 Solow 模型中的核心方程。

5.  **用图理解“稳态”资本存量**$k_{ss}$。 为了理解序列 $k_{t}$ 的动态，我们使用 45度图。首先需要指定$f=Ak^{\alpha}$ 并设置参数 $L=100，A=1.0、\alpha =0.3、s =0.2 和 \delta = 0.1$。

```
library(ggplot2)
#| label: fig-solow
#| fig-cap: "Locations of earthquakes on La Palma since 2017"
#| fig-alt: "A scatterplot of earthquake locations plotting latitude
#|   against longitude."
# 设置资本存量范围从 0 到 400
K <- seq(0, 400, length.out = 400)

# 输出函数（看起来像是立方根函数）
c <- 160 / (400 ** (1/3))  # 假设当资本存量是 400 时最大输出是 160
Y <- c * K ** (1/3)

# 储蓄函数（假设恒定储蓄率为输出的 1/4）
s <- 1/4
S <- s * Y

# 折旧函数（线性）
dep_rate <- S[which(K >= 250)[1]] / 250
D <- dep_rate * K

# 创建数据框以供 ggplot 使用
data <- data.frame(K, Y, S, D)

# 绘制图像
ggplot(data, aes(K)) +
  geom_line(aes(y = Y, color = "产出")) +
  geom_line(aes(y = S, color = "投资")) +
  geom_line(aes(y = D, color = "资本折旧")) +
  labs(x = "资本存量", y = "数值", title = "经济函数") +
  scale_color_manual(values = c("red", "blue", "magenta")) +
  theme_minimal() +
  theme(text = element_text(family = cnfont, size = 10),
        axis.title = element_text(size = 14),
        axis.text = element_text(size = 12),
        plot.title = element_text(size = 16, hjust = 0.5))
```

$$k_{ss}:=(\frac{s A}{\delta})^{1 /(1-\alpha)}$$ 稳态时人均资本存量仅由储蓄率 $s$、折旧率 $\delta$ 和资本的产出弹性 $\alpha$ 以及全要素生产率 $A$共同决定。

-   稳态资本存量表达式反映出：在资本存量到达稳态均衡时，资本存量会增长；而高于稳态均衡时，资本存量会减少。

-   让我们回到@fig-solow。资本存量较低时，投资大于折旧，因此资本存量增加；在资本存量较高时，投资小于折旧，因此资本存量下降。在稳态点时，投资水平刚好足以弥补折旧，使资本存量保持不变。

-   这在体育中很常见。当资本存量处于很低水平时候，如一个城市中赛事或者场地设施的仅有很少的投资，那么所有的“第一次”的资本存量就能够带来更大的实物资本的边际产量 $MP_K$。这个时候的资本存量一方面对产出边际贡献度更大一些，随之带来的是投资额度的增加，这进一步加快了资本的积累。另一方面，此时的资本折旧水平相对较低，新购买的实物资本能够带来更多产出，而不是替代、翻修、更新已有的资本。但是当资本存量积累到一定规模后，社会中需要修缮的资本比重越来越高，在报酬递减率的影响下，实物资本对生产带来的边际作用越来越小。

-   现实的数据也支持了这一点。从经验数据中表明了这一点，首先德国、日本在战后刚恢复的几年间，经济增长率（ $Y$ 的增长率）非常高，但当发展到一定阶段后，经济增长率开始下降。经历了从 high growth 到 low growth 的阶段。

1.  **进一步考虑储蓄和投资对收入的影响**。

如果生产率 $A$ 和劳动力投入 $L$ 是常数，那么储蓄率对经济的长期增长没有影响。至少在这个模型中，资本存量（因此还有产出）最终会收敛到其稳态值，经济不会增长。

因此，储蓄率的差异并不能解释我们看到的各国增长率差异。这并不是说储蓄无关紧要，但它不可能是国家绩效的主要因素。那么，如果储蓄对长期增长没有影响，它会做什么？它影响收入水平。储蓄率较高的国家具有较高的稳态资本存量，因此具有较高的稳态收入。把储蓄率提高，实质是投资曲线的上移，导致稳态均衡时的资本存量水平更高。该方式能够将 low growth 短期转换到 high growth 的轨道上。当达到新的稳态均衡后，经济增长为 0。实物资本的积累仅维持短期的经济增长。将储蓄率从 0.2 增加到 0.25，储蓄会发生什么。在资本存量的每个值，储蓄都更高。结果，稳态资本存量（虚线与折旧相交处）更高。并且由于资本较高，产出也会较高。

```
library(ggplot2)
#| label: fig-solow
#| fig-cap: "Locations of earthquakes on La Palma since 2017"
#| fig-alt: "A scatterplot of earthquake locations plotting latitude
#|   against longitude."
# 设置资本存量范围从 0 到 400
K <- seq(0, 400, length.out = 400)

# 输出函数（看起来像是立方根函数）
c <- 160 / (400 ** (1/3))  # 假设当资本存量是 400 时最大输出是 160
Y <- c * K ** (1/3)

# 储蓄函数（假设恒定储蓄率为输出的 1/4）
s <- 1/4
S <- s * Y

# 折旧函数（线性）
dep_rate <- S[which(K >= 250)[1]] / 250
D <- dep_rate * K

# 创建数据框以供 ggplot 使用
data <- data.frame(K, Y, S, D)

# 绘制图像
ggplot(data, aes(K)) +
  geom_line(aes(y = Y, color = "产出")) +
  geom_line(aes(y = S, color = "投资")) +
  geom_line(aes(y = D, color = "资本折旧")) +
  labs(x = "资本存量", y = "数值", title = "经济函数") +
  scale_color_manual(values = c("red", "blue", "magenta")) +
  theme_minimal() +
  theme(text = element_text(family = cnfont, size = 10),
        axis.title = element_text(size = 14),
        axis.text = element_text(size = 12),
        plot.title = element_text(size = 16, hjust = 0.5))
```

7.**关于对增长的分析**。 在经济增长的理论中，储蓄和投资是重要的增长因素，但它们并不是唯一的。储蓄和投资有助于资本积累，但若没有其他因素的参与，它们只能带来短期的增长而非长期的。要实现持续的增长，特别是在人均产出的层面上，其他两个关键因素就显得尤为重要：劳动力增长和全要素生产率（TFP）增长。以下是这两个因素在增长中的作用：

-   **劳动力增长**：增加劳动力可以提高生产总量，但在人均产出层面，它只有在结合生产率改进时才能贡献增长。即使劳动力数量增加，如果每个劳动者的生产率没有提高，人均产出可能不会增长。

-   **全要素生产率增长**：这是衡量生产效率和技术进步的指标。TFP增长意味着用相同的投入（资本和劳动）可以生产更多的产出，或者用更少的投入生产相同的产出。它反映了技术改进、管理能力提升、技能提高、创新和制度优化等因素的效果。

在增长分析中，我们还需要注意一个**平衡增长路径**（Balanced Growth Path, BGP）的概念，指一个经济体中产出和资本以相同的恒定速率增长。在平衡增长路径的假设，是基于卡尔多事实 3（随着时间的推移，资本与产出的比例大致保持稳定）。只有在平衡增长路径上，资本产出比才接近一个常数。这也是为什么 Solow 模型要分析**稳态资本产出比**的原因，最后我们将发现满足资本产出比接近常数的前提是资本与产出的增长率相同。注意，资本产出比，现实中经常使用**资本密集度capital intensity**（以资本产出比衡量）的概念。

对应现实中的例子：中国投资了惊人的 40% 的 GDP。这是否太多了？关于中国的投资率是否过高的问题，Solow 的观点是：取决于其经济增长的速度和全要素生产率的提高。如果中国的经济增长速度非常快，那么高投资率是有道理的，因为它需要这些投资来支撑和维持快速的增长。资本产出比的高低并不仅仅取决于投资的绝对量，而且取决于这些投资在经济中的有效性，以及它们是如何与增长率相结合的。

另一个例子是：德国和日本战后初期的高增长阶段说明了资本积累对经济增长的推动作用，而后随着资本存量的增大，增长率开始下降，反映了报酬递减的经济效应。这在稳态均衡中已经有所展示。

**静态与动态分析**:提高储蓄率将导致资本存量的稳态水平提高，从而短期内推动经济从低增长过渡到高增长，但长期看，稳态下经济增长率趋于零，依赖于资本积累的增长无法持续。

$$
\begin{aligned}
L_{t+1} & =(1+n) L_t \\ A_{t+1} & =(1+a) A_t 
\end{aligned}
$$ 产出和资本增长有多快？在经济学中，我们常常假设产出和资本以相同的速率增长，这是因为我们在考虑一个**平衡增长路径（Balanced Growth Path, BGP）**。在这个路径上，所有宏观经济变量都以相同的恒定速率增长，这使得模型简化，并且易于分析。这种假设背后的直觉是，如果资本和产出增长速度不同，那么资本和产出之间的比率将会不断变化，这与长期增长的稳定性假设相矛盾。我们猜测产出的增长速度（ $g$）和资本存量增长的速率增长是相同的。从生产函数中，可知：

$$
\begin{aligned}
(1+g)& =\begin{array}{c}Y_{t+1}/Y_t\end{array}  \\
&=\quad(A_{t+1}/A_t)(K_{t+1}/K_t)^\alpha(L_{t+1}/L_t)^{1-\alpha} \\
&=\quad(1+a)(1+g)^\alpha(1+n)^{1-\alpha}
\end{aligned}$$

经济增长率为： $$\begin{array}
{rcl}(1+g) = (1+a)^{1/(1-\alpha)}(1+n)
\end{array}$$

上式透露重要的信息，它表明**产出增长与全要素生产率和劳动力的增长率相关**。再次注意，储蓄率不会影响这种增长率。同样，每位工人的产出增长率是： $$\begin{array}{rcl}(1+g)/(1+n)&=&(1+a)^{1/(1-\alpha)}\end{array}$$

仅取决于全要素生产率的增长。如果 $a > 0$ ，则每位工人的产出增长率高于全要素生产率的增长率，因为指数 $1/(1 − \alpha) > 1$。实际的含义是：全要素生产率对产出的直接影响被资本存量的增长放大；这与我们之前做出的一个备注相符：**资本积累往往会强化生产率增长的影响**。

对于资本存量而言，储蓄率高的国家是否拥有相对于其经济规模而言更多的资本？答案是肯定的，但产出快速增长的国家也确实需要大量投资才能维持资本与产出的恒定比率。

中国虽然投资率很高（资本存量增长的快），但它需要维持与美国相当的资本产出比率。再次考虑资本和产出以相同速率 $g$ 增长的“稳态”。为了说明上述观点，将资本存量方程变为： $$\begin{aligned}K_{t+1}&=\quad(1-\delta)K_t+sY_t\\(1+g)K_t/Y_t&=\quad(1-\delta)K_t/Y_t+s\end{aligned}$$

求解 $K/Y$ 给我们稳态资本产出比： $$(K/Y)_{ss}\quad=\quad\left(\frac s{\delta+g}\right)$$

您可能会注意到，这一论证实际上并不需要索洛模型：我们所需要的只是将今天的资本与明天的资本联系起来的方程式。如果我们将资本存量方程除以 $Y_t$ 并代入 $Y_t=Y_{t+1}/(1+g)$ 可得： $$(1+g)K_{t+1}/Y_{t+1}=(1-\delta)K_t/Y_t+s$$ 稳态资本产出比（使 Kt+1/Yt+1 和 Kt/Yt 相等的比率）正是我们刚刚推导出来的。

为了回到理解资本密集度来源的目标，请注意增长对稳态资本产出比的影响。对于给定的储蓄/投资率$s$，产出增长率较高的国家，每单位产出中资本占比相对较少。为什么？因为产出增长很快，你需要大量投资才能使资本与产出保持一致。

**例子**。以下是根据美国情况得出的部分数字：人口增长率n = 0.005 (0.5%)，全要素增长率 $a = 0.01 (1%)$，储蓄率$s = 0.15$，折旧率 $\delta=0.06$。产出的增长率是多少？稳态资本产出比是多少？增长率满足:

$$1+g=(1+a)^{1/(1-\alpha)}(1+n)=1.015\times1.005=1.0201$$ 在这里，与常用的资本对产出弹性的惯例使用 $\alpha = 1/3$。使用与我们之前相同的参数，得到稳态资本产出比率为

$$1.872~=~\frac{0.15}{0.06+0.020}$$ 现在考虑基于中国的数字：人口增长率$n = 0.005 (0.5%)$ 和折旧率 $\delta=0.06$与美国一样，但全要素增长率 $a = 0.06 (6%)$ 储蓄率$s = 0.30$。重新计算可得：增长率现在为 $g = 0.0968$，资本产出比降至 $1.013$。请注意，尽管中国的储蓄/投资率大幅增加，但增长却很小。中国储蓄过多吗？也许没有：他们的资本产出比（通过此计算）与美国的没有太大不同。中国需要大量投资，仅仅是为了跟上经济增长。
